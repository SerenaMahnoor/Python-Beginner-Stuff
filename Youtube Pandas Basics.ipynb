When rearranging data, always print before assigning! like so
~~~
print (poke.iloc[:, 4:10].sum(axis=1))
~~~

that way you can check the outcome of a code before fucking up the data itself

```python
#add a column that builds on input from other columns
poke['Total']= poke['HP'] + poke['Attack']+poke['Defense']+poke['Sp. Atk']+poke['Sp. Def']+poke['Speed']
poke.sort_values('Total',ascending=False)[:5]
#remove that column
poke = poke.drop(columns=['Total']) #needs to be equated to the df to alter it directly
#another way to do it to sum the specified column items vertically
#we specify axis=1 to make it sum vertically, default is horizontal.
poke['Total'] = poke.iloc[:, 4:10].sum(axis=1)
poke.head()
```

```python
#define the columns as an object to be handled
cols = list(poke.columns.values)
#manually rearrange columns 0-3, then the last one, then the rest.
#the single column counts as a string thats why it has to be added within another set of block brackets
poke = poke[cols[0:4] + [cols [-1]] +cols[4:12]]
poke.head()
```

## Advanced Stuff- Filtering Data- 31 minute onwards

```python
poke.head()
```

```python
#I want to find all entries with two specific items in seperate columns
#before, i did this:
#print(poke.loc[poke['Type 1']=="Fire"][['Name','Sp. Def']][0:5])
#writing "and" wont work!! convention is and signal
print(poke.loc[(poke['Type 1'] == "Grass") & (poke["Type 2"] == "Poison") ][:5])

#write or sign "|" if I want all entries that are either of the two specified above
poke.loc[(poke['Type 1'] == "Grass") | (poke["Type 2"] == "Poison") ]
```

```python
#Can also be with numbers, and can assign that as a seperate dataframe
grass_boss = poke.loc[(poke['Type 1'] == "Grass") & (poke["Type 2"] == "Poison") & (poke["HP"] > 70) ]
print(grass_boss)
#note that the old index is still there- might be annoying in certain cases
print(grass_boss.reset_index(drop= True, inplace= True))
grass_boss
#and old index gone!
```

```python
#Let's remove all names with Mega
#the squiggly line returns all that aren't what has been specified
#Plus I can mention that of the non-Megas, I only want the Grass and Poison. 
poke.loc[(~poke['Name'].str.contains('Mega')) & (poke['Type 1'] == "Grass") | (poke['Type 1'] == "Fire") ]
```

```python
#Can filter with regular expressions
import re
#no spaces between the "|" sign cuz then it expects the entry to have a space in it..
#also, in case my entries are mixed capital and lower case,
#I add "flags=re.I"
poke.loc[poke['Type 1'].str.contains('Fire|grass', flags=re.I, regex=True)]
```

```python
#Let's get all pokemon with names starting with "Pi"
poke.loc[poke["Name"].str.contains("Pi")]
```

```python
#the reason I get all names with pi in it here is cuz of capital letter ignore
poke.loc[poke['Name'].str.contains('pi', flags=re.I, regex=True)]

#to get only the ones STARTING with "pi", write "^" before it
poke.loc[poke['Name'].str.contains('^pi', flags=re.I, regex=True)]
```

## Conditional Changes
We can change our dataframes based on the conditions we filter out by

```python
#Say I wana change the Type 1 type Fire to Hotman
poke.loc[poke["Type 1"] == "Flameo", "Type 1"] = "Hotman"
poke.loc[poke["Type 1"] == "Hotman"][:5]
```

This is VERY tricky, cuz if you alter something irreversibly, you need to have a checkpoint saved- like xlsx, to save from again. 

```python
#alas, all good things must come to an end
#poke.loc[poke["Type 1"] == "Fire", "Type 1"] = "Hotman"
poke.loc[poke["Type 1"] == "Hotman", "Type 1"] = "Fire"
poke.loc[poke["Type 1"] == "Fire"][:5]
#here, the ", Type 1" after the hotman is telling where I want a change to be made.
#For Type 1 equals Fire, Type 1 should change to Hotman, and v.v for
```

```python
poke.loc[poke["Total"] > 500, ["Generation", "Legendary"]] = ["Epico Generation", "Extra Legendary"]
poke.loc[poke["Total"] > 500]
```

## Aggregate Statistics (Groupby)


```python
#to see the vg hp and attack for each type 1 group
poke.groupby(["Type 1"]).mean().sort_values("Speed", ascending= False)#fuck cool
# So Steel pokemon have highest defense, Dragon has highest attack, Flying has highest speed.
```

```python
print(poke.groupby(["Type 1"]).count()[:3])
print(poke.groupby(["Type 1"]).sum()[:3])
```

```python
counts = poke.groupby(["Type 1", "Type 2"]).count()
counts["#"]
```
